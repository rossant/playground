Computing all pairwise correlograms among a set of spike trains.
GPGPU implementation with PyOpenCL.

```python
import numpy as np
import pyopencl as cl
```

```python
import os
os.environ['PYOPENCL_COMPILER_OUTPUT'] = '1'
```

```python
platform = cl.get_platforms()
my_gpu_devices = platform[1].get_devices()
my_gpu_devices
```

```python
ctx = cl.Context(devices=my_gpu_devices)
```

```python
queue = cl.CommandQueue(ctx)
```

```python
nspikes = 1000000
nneurons = 100
sampling_rate = 20000
```

```python
spiketimes = (np.cumsum(np.random.exponential(scale=.005, size=nspikes))*
              sampling_rate).astype(np.int64)
neurons = np.random.randint(low=0, high=nneurons, size=nspikes).astype(np.int32)
duration = np.int64(spiketimes[-1])
```

```python
dt = np.int32(40)  # 1 ms bin
T = np.int32(20*50)  # 50ms half window
```

```python
neurons_unique = np.unique(neurons)
```

```python
mf = cl.mem_flags
spiketimes_g = cl.Buffer(ctx, mf.READ_ONLY | mf.COPY_HOST_PTR, hostbuf=spiketimes)
neurons_g = cl.Buffer(ctx, mf.READ_ONLY | mf.COPY_HOST_PTR, hostbuf=neurons)
```

```python
prg = cl.Program(ctx, """
__kernel void correlograms(__global const long *spiketimes_g,
                           __global const int *neurons_g,
                           __global int *corr_g,
                           long duration, int T, int dt, int nneurons) {
  
  int nspikes = get_global_size(0);
  int d = 0;
  int Toverdt = T/dt;
  int idx0, c0, idx1, c1, k;
  long t0, t1, t0min, t0max;
  
  idx0 = get_global_id(0);
  t0 = spiketimes_g[idx0];
  t0min = t0 - T;
  t0max = t0 + T;
  c0 = neurons_g[idx0];
  
  
  for (k=1; k<nspikes-idx0; k++) {
      idx1 = idx0 + k;
      t1 = spiketimes_g[idx1];
      c1 = neurons_g[idx1];
      
      if (t1 > t0max) break;
      
      d = (t1-t0) / dt;
      atomic_add(&corr_g[(c0*nneurons*(2*Toverdt+1))+
                         (c1*(2*Toverdt+1))+
                         (d+Toverdt)], 1);
      
  }
  
  for (k=-1; k>=-idx0; k--) {
      idx1 = idx0 + k;
      t1 = spiketimes_g[idx1];
      c1 = neurons_g[idx1];
      
      if (t1 < t0min) break;
      
      d = (t1-t0) / dt;
      atomic_add(&corr_g[(c0*nneurons*(2*Toverdt+1))+
                         (c1*(2*Toverdt+1))+
                         (d+Toverdt)], 1);
      
  }
  
}
""").build()
```

```python
corr = np.zeros((nneurons, nneurons, 2*(T//dt)+1), 
                dtype=np.int32)
corr_g = cl.Buffer(ctx, mf.WRITE_ONLY, corr.nbytes)
cl.enqueue_copy(queue, corr_g, corr);
```

```python
prg.correlograms(queue, (nspikes,), None, 
                 spiketimes_g, neurons_g, corr_g, duration,
                 T, dt, np.int32(nneurons))
cl.enqueue_copy(queue, corr, corr_g);
```

```python
%%timeit
prg.correlograms(queue, (nspikes,), None, 
                 spiketimes_g, neurons_g, corr_g, duration,
                 T, dt, np.int32(nneurons))
cl.enqueue_copy(queue, corr, corr_g);
```
